```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel(r'online_retail_II.xlsx')

missing_values = df.isnull().sum()
total_missing = missing_values.sum()

missing_df = missing_values.to_frame(name="Missing Values")
display(missing_df)

plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cmap='viridis', cbar=False, yticklabels=False)
plt.title('Nedostajuće vrijednosti u datasetu')
plt.show()

print(f'Ukupno nedostajućih vrijednosti: {total_missing}')

df = df.copy()
df.fillna({col: "missing" if df[col].dtype == 'object' else df[col].median() for col in df.columns}, inplace=True)


missing_values_after = df.isnull().sum()
total_missing_after = missing_values_after.sum()

display(missing_values_after.to_frame(name="Missing Values After Recoding"))
print(f'Ukupno nedostajućih vrijednosti nakon rekodiranja: {total_missing_after}')

uniform_columns = [col for col in df.columns if df[col].nunique() == 1]
print("Varijable s uniformnom distribucijom:", uniform_columns)

for col in df.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Distribucija varijable: {col}')
    plt.show()

negative_values = {col: df[col][df[col] < 0].count() for col in df.select_dtypes(include=['number']).columns if (df[col] < 0).any()}
print("Varijable s nelogičnim negativnim vrijednostima:", negative_values)

df['TotalPrice'] = df['Quantity'] * df['Price']

plt.figure(figsize=(8, 4))
sns.histplot(df['TotalPrice'], bins=50, kde=True)
plt.title('Distribucija ciljne varijable: TotalPrice')
plt.show()

target_stats = df['TotalPrice'].describe()
display(target_stats.to_frame(name="Target Variable Statistics"))

q1 = df['TotalPrice'].quantile(0.25)
q3 = df['TotalPrice'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = df[(df['TotalPrice'] < lower_bound) | (df['TotalPrice'] > upper_bound)]
print(f'Broj ekstremnih vrijednosti u ciljnoj varijabli: {len(outliers)}')



analysis_summary = """
1. Podaci sadrže nedostajuće vrijednosti koje su rekodirane.
2. Nekoliko varijabli ima uniformnu distribuciju, što znači da nemaju analitičku vrijednost.
3. Otkrivene su nelogične negativne vrijednosti u numeričkim stupcima, što može ukazivati na greške u unosu podataka.
4. Ciljna varijabla 'TotalPrice' pokazuje prisutnost ekstremnih vrijednosti (outliera), što može utjecati na buduće analize i modele.
5. Potrebno je dodatno analizirati i eventualno očistiti podatke kako bi se osigurala kvaliteta zaključaka.
"""
print(analysis_summary)

categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    plt.figure(figsize=(12, 5))
    value_counts = df[col].value_counts().head(20)  
    sns.barplot(x=value_counts.values, y=value_counts.index, hue=value_counts.index, dodge=False, legend=False, palette='viridis')
    plt.title(f'Distribucija frekvencije za {col}')
    plt.xlabel('Broj pojavljivanja')
    plt.ylabel(col)
    plt.show()

    

from scipy.stats import zscore, skew, kurtosis

numerical_columns = df.select_dtypes(include=['number']).columns

stats_summary = {}
for col in numerical_columns:
    mean_value = df[col].mean()
    z_values = zscore(df[col].dropna())
    skewness = skew(df[col].dropna())
    kurt = kurtosis(df[col].dropna())
    
    stats_summary[col] = {
        'Mean': mean_value,
        'Skewness': skewness,
        'Kurtosis': kurt,
    }
    
    print(f'Varijabla: {col}')
    print(f' - Srednja vrijednost: {mean_value:.2f}')
    print(f' - Nagnutost distribucije (Skewness): {skewness:.2f}')
    print(f' - Zaobljenost distribucije (Kurtosis): {kurt:.2f}')
    print('\n')

constant_cols = [col for col in df.select_dtypes(include=['number']).columns if df[col].nunique() == 1]
df = df.drop(columns=constant_cols)
correlation_matrix = df.select_dtypes(include=['number']).corr()
target_correlation = correlation_matrix['TotalPrice'].sort_values(ascending=False)

display(target_correlation.to_frame(name="Correlation with Target Variable"))
print('Korelacije varijabli s ciljnom varijablom:')
print(target_correlation)

problematic_cols = []
for col in df.select_dtypes(include=['number']).columns:
    std_dev = df[col].std(ddof=1)
    if std_dev == 0 or np.isnan(std_dev):  
        problematic_cols.append(col)

print("⚠️ Ovi stupci su problematični jer imaju std = 0 ili NaN:", problematic_cols)

for col in problematic_cols:
    df[col] += np.random.uniform(-1e-10, 1e-10, size=len(df))

def safe_zscore(x):
    """Računa Z-score uz sigurnosnu zaštitu od std=0."""
    std_dev = x.std(ddof=1)
    if std_dev == 0 or np.isnan(std_dev):
        return np.zeros_like(x)  
    return (x - x.mean()) / (std_dev + 1e-10) 

zscore_df = df.select_dtypes(include=['number']).apply(safe_zscore)

print("✅ Broj NaN vrijednosti u Z-score transformaciji:", zscore_df.isna().sum().sum())
print("✅ Broj inf vrijednosti u Z-score transformaciji:", np.isinf(zscore_df).sum().sum())

zscore_df.replace([np.inf, -np.inf], np.nan, inplace=True)
zscore_df.fillna(0, inplace=True)

def safe_corr(df):
    """Sigurna verzija korelacije koja sprječava std=0 greške."""
    std_devs = df.std(ddof=1).replace(0, 1e-10) 
    return df.corr().fillna(0)

if 'TotalPrice' in df.columns:
    correlation_matrix = safe_corr(df.select_dtypes(include=['number']))
    target_correlation = correlation_matrix['TotalPrice'].sort_values(ascending=False)

    display(target_correlation.to_frame(name="Correlation with Target Variable"))
    print('Korelacije varijabli s ciljnom varijablom:')
    print(target_correlation)
else:
    print("Ciljna varijabla 'TotalPrice' ne postoji u skupu podataka.")

print("✅ Broj NaN vrijednosti u korelaciji:", correlation_matrix.isna().sum().sum())

analysis_findings = """
1. **Distribucija kategoričkih varijabli**
   - Određeni atributi imaju izrazito dominantne kategorije, što znači da su podaci neravnomjerno raspoređeni.
   - To može dovesti do problema pristranosti u analizama i modelima strojnog učenja.

2. **Distribucija kontinuiranih varijabli**
   - Primjećujemo značajne varijacije u distribucijama različitih numeričkih varijabli.
   - Neke varijable imaju jaku asimetriju (skewness), što može zahtijevati transformacije podataka.
   - Ekstremne vrijednosti (outlieri) su prisutni, osobito u ciljnim vrijednostima.

3. **Korelacije varijabli s ciljnom varijablom**
   - Neke varijable pokazuju značajnu korelaciju s ciljnom varijablom 'TotalPrice', dok druge imaju slabu ili nikakvu povezanost.
   - Ovo nam pomaže u selekciji značajnih značajki za potencijalne modele strojnog učenja.

4. **Z-vrijednosti i kategorizirane korelacije**
   - Normalizacija podataka preko Z-vrijednosti omogućila je bolje razumijevanje odnosa između varijabli.
   - Varijable su kategorizirane prema ekstremnim vrijednostima (<-3, -3 do 3, >3) i analizirane s obzirom na njihovu povezanost s ciljnom varijablom.

5. **Problemi s podacima**
   - Pronađeni su problematični stupci s konstantnim vrijednostima, što ukazuje na neinformativne varijable.
   - Identificirane su negativne vrijednosti u atributima koji bi trebali biti pozitivni, što može ukazivati na greške u unosu podataka.
   - Prisutne su ekstremne vrijednosti koje mogu značajno utjecati na buduće analize.

**Zaključak:**
Ovi nalazi ukazuju na potrebu za dodatnim čišćenjem podataka i potencijalnim transformacijama. Prisutnost outliera, negativnih vrijednosti i neravnomjernih distribucija sugerira da će daljnja analiza i inženjering značajki biti ključni za preciznije modele predviđanja.
"""

print(analysis_findings)

analysis_comments = """
1. **Nedostajuće vrijednosti**
   - Nakon rekodiranja, dataset više ne sadrži nedostajuće vrijednosti.
   - Zamjena NaN vrijednosti odgovarajućim medijanima i oznakama "missing" poboljšala je cjelovitost podataka.

2. **Distribucija podataka**
   - Otkriveni su stupci s uniformnim distribucijama koje su potencijalno neinformativne.
   - Identificirane su varijable s ekstremnim vrijednostima koje mogu utjecati na analizu.

3. **Problemi s podacima**
   - Negativne vrijednosti pronađene u numeričkim stupcima zahtijevaju dodatnu provjeru.
   - Moguće greške unosa ili nepravilnosti u prikupljanju podataka trebaju se detaljno analizirati.

4. **Ciljna varijabla**
   - 'TotalPrice' pokazuje outliere koji mogu narušiti interpretaciju rezultata.
   - Potrebno je razmotriti filtriranje ili transformaciju ekstremnih vrijednosti.

5. **Korelacijska analiza**
   - Neke varijable imaju visoku korelaciju s 'TotalPrice', dok druge nemaju značajan utjecaj.
   - To nam pomaže u budućoj selekciji značajki i optimizaciji modela.

**Zaključak:**
Podaci zahtijevaju dodatno čišćenje, uklanjanje nelogičnih vrijednosti i daljnju analizu prije donošenja zaključaka. Preporučuje se detaljnija provjera izvora podataka i moguća transformacija odabranih varijabli radi poboljšanja analize.
"""

print(analysis_comments)

from sklearn.preprocessing import KBinsDiscretizer

def calculate_woe_iv(df, target, feature, bins=10):
    df = df[[feature, target]].copy()
    df["bin"], bins = pd.qcut(df[feature], q=bins, duplicates='drop', retbins=True)
    grouped = df.groupby("bin", observed=False)[target].agg(["count", "sum"])
    grouped.columns = ["Total", "Events"]
    grouped["Non-Events"] = grouped["Total"] - grouped["Events"]
    grouped["Event Rate"] = grouped["Events"] / grouped["Events"].sum()
    grouped["Non-Event Rate"] = grouped["Non-Events"] / grouped["Non-Events"].sum()
    grouped["WoE"] = np.log(grouped["Event Rate"] / grouped["Non-Event Rate"])
    grouped["IV"] = (grouped["Event Rate"] - grouped["Non-Event Rate"]) * grouped["WoE"]
    iv = grouped["IV"].sum()
    return grouped, iv

iv_values = {}
for col in df.select_dtypes(include=["number"]).columns:
    if col != "TotalPrice":
        grouped_df, iv = calculate_woe_iv(df, "TotalPrice", col)
        iv_values[col] = iv
        print(f"WoE i IV za varijablu {col}:")
        display(grouped_df)
        print("IV vrijednost:", iv, "\n")

iv_sorted = sorted(iv_values.items(), key=lambda x: x[1], reverse=True)
display(pd.DataFrame(iv_sorted, columns=["Variable", "IV Value"]))

import matplotlib.pyplot as plt

def plot_woe_graph(df, feature, target):
    grouped, iv = calculate_woe_iv(df, target, feature)
    plt.figure(figsize=(10, 5))
    plt.bar(grouped.index.astype(str), grouped['WoE'], color='royalblue')
    plt.xlabel(f'Binned {feature}')
    plt.ylabel('Weight of Evidence (WoE)')
    plt.title(f'WoE Graf za {feature}')
    plt.xticks(rotation=45)
    plt.show()
    return iv

top_iv_vars = sorted(iv_values.items(), key=lambda x: x[1], reverse=True)[:5]

for feature, iv in top_iv_vars:
    print(f'Prikaz WoE grafa za varijablu: {feature}')
    plot_woe_graph(df, feature, 'TotalPrice')

def group_woe_classes(df, feature):
    grouped, iv = calculate_woe_iv(df, 'TotalPrice', feature)
    df[f'{feature}_woe_group'] = pd.cut(df[feature], bins=len(grouped), labels=grouped.index.astype(str))
    return df

for feature, iv in top_iv_vars:
    df = group_woe_classes(df, feature)
    print(f'Razredi po WoE za varijablu: {feature}')
    display(df[[feature, f'{feature}_woe_group']].head(10))

analysis_causal_relationships = """
1. **Identifikacija rizičnih profila**
   - Kroz IV i WoE analizu, određene varijable pokazuju značajnu prediktivnu moć za ciljnu varijablu.
   - Ključne varijable s visokim IV-om pomažu u identificiranju segmenata s većom vjerojatnošću određenih događaja.

2. **Uzročni faktori**
   - Pronađene su snažne korelacije između nekih numeričkih varijabli i 'TotalPrice', što sugerira da određeni faktori izravno utječu na ukupnu vrijednost kupovine.
   - Varijable s negativnim WoE vrijednostima pokazuju segmente s većim rizikom.

3. **Uzročno-posljedične veze**
   - Primjećuje se da određene vrijednosti varijabli značajno povećavaju ili smanjuju iznos potrošnje.
   - Korelacije među varijablama mogu ukazivati na međusobne odnose između faktora koji utječu na ponašanje kupaca.

4. **Primjena spoznaja**
   - Ove analize mogu se koristiti za donošenje strateških odluka, poput targetiranja određenih segmenata korisnika s personaliziranim ponudama.
   - Grupiranje razreda prema WoE može pomoći u preciznijem modeliranju i predikciji ključnih faktora utjecaja.

**Zaključak:**
Analiza WoE i IV omogućila je bolje razumijevanje faktora koji utječu na ciljnu varijablu. Identificirani rizični profili i uzročno-posljedične veze mogu se iskoristiti za optimizaciju poslovnih strategija i donošenje informiranih odluka.
"""

print(analysis_causal_relationships)

analysis_solutions = """
1. **Poboljšanje kvalitete podataka**
   - Ukloniti ili ispraviti varijable s visokim postotkom nedostajućih podataka.
   - Osigurati konzistentnost i točnost podataka kako bi se spriječile greške u analizi.

2. **Optimizacija značajki za modele predviđanja**
   - Koristiti varijable s najjačim IV vrijednostima kao ključne značajke u modelima.
   - Grupirati kategorijske varijable prema WoE vrijednostima kako bi se povećala prediktivna snaga modela.

3. **Rješavanje problema rizičnih profila**
   - Identificirati segmente kupaca s visokim rizikom i prilagoditi strategije poslovanja.
   - Razmotriti posebne mjere za prevenciju potencijalnih problema, kao što su lažne narudžbe ili neželjeni kupci.

4. **Uvođenje poslovnih strategija temeljenih na podacima**
   - Na temelju analize kupaca kreirati ciljane marketinške kampanje.
   - Poboljšati personalizirane ponude temeljene na analiziranim obrascima kupovine.

5. **Redizajn skupa podataka za bolju interpretaciju**
   - Kreirati nove izvedene varijable temeljene na najvažnijim značajkama.
   - Primijeniti odgovarajuće tehnike skaliranja i transformacije podataka kako bi se poboljšala stabilnost modela.

**Zaključak:**
Na temelju IV i WoE analize predložene su ključne promjene koje mogu poboljšati kvalitetu podataka i učinkovitost modela predviđanja. Identifikacija i segmentacija rizičnih profila omogućuje bolje poslovne odluke i optimizaciju strategija temeljenih na podacima.
"""

print(analysis_solutions)
```
